```python
# Import necessary libraries
import numpy as np
import os
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from tensorflow.keras.preprocessing.image import img_to_array

# Assuming you have loaded and preprocessed your dataset
# Define paths and labels (adjust as needed for your dataset)
train_dir = '/Users/admin/.cache/kagglehub/datasets/asdasdasasdas/garbage-classification/versions/2/Garbage classification/Garbage classification'
labels = ['cardboard', 'glass', 'metal', 'paper', 'plastic', 'trash']

# Load images and prepare your data (X_train, y_train, X_test, y_test)
# You can resize images, flatten them, and create your training/testing sets.
# Below is a simple example of loading and flattening images for logistic regression:

X_data = []
y_data = []

# Loop through labels and images
for label_index, label in enumerate(labels):
    directory = os.path.join(train_dir, label)
    for image_name in os.listdir(directory):
        img_path = os.path.join(directory, image_name)
        img = mpimg.imread(img_path)
        
        # Resize image to a consistent shape (e.g., 64x64 pixels) and flatten it
        img_resized = np.resize(img, (64, 64, 3))  # Resize to 64x64, 3 channels (RGB)
        img_array = img_to_array(img_resized).flatten()  # Flatten the image into 1D array
        
        X_data.append(img_array)  # Add to feature list
        y_data.append(label_index)  # Add corresponding label index

# Convert lists to numpy arrays for processing
X_data = np.array(X_data)
y_data = np.array(y_data)

# Split into training and testing datasets (e.g., 80% training, 20% testing)
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_data, y_data, test_size=0.2, random_state=42)

# Initialize and train the Logistic Regression model
logreg = LogisticRegression(max_iter=1000)
logreg.fit(X_train, y_train)

# Make predictions on the test data
y_pred_logreg = logreg.predict(X_test)

# Calculate accuracy of the Logistic Regression model
logreg_accuracy = accuracy_score(y_test, y_pred_logreg)

# Print the accuracy in percentage
print(f'Logistic Regression Accuracy: {logreg_accuracy * 100:.2f}%')
```
